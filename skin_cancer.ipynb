```python
import os
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Set up paths
IMAGE_PATH = r"C:\Users\Shadow\Downloads\archive\all_images"
METADATA_PATH = r"C:\Users\Shadow\Downloads\archive\HAM10000_metadata.csv"

# Load metadata
metadata = pd.read_csv(METADATA_PATH)

# Display first few rows and info about the metadata
print(metadata.head())
print(metadata.info())

# Check the distribution of diagnoses
print(metadata['dx'].value_counts(normalize=True))

# Function to load and preprocess image
def load_image(image_id):
    img_path = os.path.join(IMAGE_PATH, image_id + '.jpg')
    img = Image.open(img_path)
    img = img.resize((224, 224))  # Resize to match typical CNN input
    return np.array(img) / 255.0  # Normalize pixel values

# Load images and create labels
images = []
labels = []
for idx, row in metadata.iterrows():
    img = load_image(row['image_id'])
    images.append(img)
    labels.append(row['dx'])
    if idx % 1000 == 0:
        print(f"Processed {idx} images")

images = np.array(images)
labels = np.array(labels)

# Encode labels
le = LabelEncoder()
labels_encoded = le.fit_transform(labels)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(images, labels_encoded, test_size=0.2, random_state=42)

print("Shape of training data:", X_train.shape)
print("Shape of testing data:", X_test.shape)

# Display a few sample images
plt.figure(figsize=(20,10))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(X_train[i])
    plt.title(le.inverse_transform([y_train[i]])[0])
    plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Define the number of classes
num_classes = len(le.classes_)

# Load the pre-trained ResNet50 model
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(num_classes, activation='softmax')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32
)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_accuracy:.4f}")

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Function to predict on new images
def predict_skin_cancer(image_path):
    img = Image.open(image_path)
    img = img.resize((224, 224))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    predictions = model.predict(img_array)
    predicted_class = le.inverse_transform([np.argmax(predictions[0])])[0]
    confidence = np.max(predictions[0])
    
    return predicted_class, confidence

# Example usage
# example_image_path = "path_to_your_test_image.jpg"
# predicted_class, confidence = predict_skin_cancer(example_image_path)
# print(f"Predicted class: {predicted_class}")
# print(f"Confidence: {confidence:.2f}")
```

```python
# Make sure we have the necessary imports
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Redefine the predict_skin_cancer function to ensure it's using our trained model
def predict_skin_cancer(image_array):
    # Ensure the image is in the right shape
    if image_array.shape != (224, 224, 3):
        raise ValueError("Image must be 224x224 pixels with 3 color channels")
    
    # Add batch dimension
    img_array = np.expand_dims(image_array, axis=0)
    
    # Make prediction
    predictions = model.predict(img_array)
    predicted_class = le.inverse_transform([np.argmax(predictions[0])])[0]
    confidence = np.max(predictions[0])
    
    return predicted_class, confidence

# Test on 5 random images from the test set
np.random.seed(120)  # for reproducibility
test_indices = np.random.choice(len(X_test), 5, replace=False)

for i, idx in enumerate(test_indices):
    img = X_test[idx]
    true_label = le.inverse_transform([y_test[idx]])[0]
    
    predicted_class, confidence = predict_skin_cancer(img)
    
    print(f"Image {i+1}:")
    print(f"True label: {true_label}")
    print(f"Predicted class: {predicted_class}")
    print(f"Confidence: {confidence:.2f}")
    print()
    
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f"True: {true_label}\nPredicted: {predicted_class}\nConfidence: {confidence:.2f}")
    plt.axis('off')
    plt.show()
```
